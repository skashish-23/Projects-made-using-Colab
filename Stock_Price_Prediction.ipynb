```python
import yfinance as yf
import pandas as pd
```

```python
bse = yf.Ticker("^BSESN")
bse = bse.history(period="max")
```

```python
del bse["Dividends"]
del bse["Stock Splits"]
```

```python
bse["Tomorrow"] = bse["Close"].shift(-1)
```

```python
bse['Target'] = (bse['Tomorrow'] > bse['Close']).astype(int)
```

```python
bse
```

```python
from sklearn.ensemble import RandomForestClassifier


train = bse.iloc[:-100]
test = bse.iloc[-100:]

predictors = ['Close', 'Volume', 'High', 'Low', 'Open']
```

```python
model = RandomForestClassifier(n_estimators=135, min_samples_split=150, random_state=1)
model.fit(train[predictors], train["Target"])
```

```python
preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
```
